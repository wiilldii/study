Приветствуем в нашем боте "Магический лес"

Данный интерактивный бот симулирует игру, где ваша цель с помощью правильных выборов добраться до сокровищ, и не умереть!

История начинается с того, что пользователь оказывается в магическом лесу, где ему нужно выбрать, как продолжить свой путь: исследовать пещеру, идти к реке или подняться на гору. Каждый выбор ведет к уникальной развилке и конечному результату.


Он основан на if/elif/else, где каждый введенный человеком параметр влияет на игру.

```python
def adventure_bot():
    print("Приветствую тебя, герой! Ты попал в магический лес, полный загадок и приключений.")
    print("Твоя цель - выбрать свой путь и найти сокровище. Помни: каждый выбор имеет последствия!")

    print("\nТы видишь три пути перед собой: \n1 - Пойти в темную пещеру. \n2 - Отправиться к шумной реке. \n3 - Взобраться на гору.")
    choice1 = input("Куда ты пойдешь? (1/2/3): ")

    if choice1 == "1":
        print("\nТы вошел в пещеру. Внутри темно, но ты находишь сундук и старый факел.")
        print("1 - Открыть сундук. \n2 - Зажечь факел и осмотреться. \n3 - Уйти из пещеры, не рискуя.")
        choice2 = input("Что ты сделаешь? (1/2/3): ")

        if choice2 == "1":
            print("\nСундук был ловушкой! Ты активировал древнее заклинание и оказался в ловушке.")
            print("Конец игры!")
        elif choice2 == "2":
            print("\nТы зажег факел и осмотрелся. Ты нашел тайный проход, ведущий глубже в пещеру.")
            print("1 - Идти по тайному проходу. \n2 - Вернуться назад.")
            choice3 = input("Что ты сделаешь? (1/2): ")

            if choice3 == "1":
                print("\nТы нашел древний артефакт, который дает тебе силу и мудрость. С этим артефактом ты легко находишь сокровище!")
                print("Поздравляю, ты победил!")
            else:
                print("\nТы решил вернуться назад и выйти из пещеры. Приключение закончено.")
                print("Конец игры!")
        else:
            print("\nТы решил не рисковать и вернулся в лес. Умное решение! Но приключение закончено.")
            print("Конец игры!")

    elif choice1 == "2":
        print("\nТы подошел к реке. Течение быстрое, но в воде ты замечаешь блестящий объект и старый мост.")
        print("1 - Попробовать достать объект. \n2 - Перейти реку по мосту. \n3 - Перейти реку вброд.")
        choice2 = input("Что ты сделаешь? (1/2/3): ")

        if choice2 == "1":
            print("\nТы достал блестящий объект - это древний амулет! Он дает тебе защиту и силы.")
            print("С таким амулетом ты легко находишь сокровище и становишься героем своего мира!")
            print("Поздравляю, ты победил!")
        elif choice2 == "2":
            print("\nТы перешел реку по мосту и нашел тайный сад с магическими растениями.")
            print("1 - Исследовать сад. \n2 - Продолжить путь.")
            choice3 = input("Что ты сделаешь? (1/2): ")

            if choice3 == "1":
                print("\nТы нашел редкое растение, которое исцеляет все болезни. С этим растением ты спасаешь своего народа от эпидемии!")
                print("Поздравляю, ты победил!")
            else:
                print("\nТы продолжил путь и нашел сокровище!")
                print("Поздравляю, ты победил!")
        else:
            print("\nТы попытался перейти реку, но течение оказалось слишком сильным. Ты потерялся в бурлящей воде.")
            print("Конец игры!")

    elif choice1 == "3":
        print("\nТы поднялся на гору и оказался у древнего храма. Ты также видишь старую хижину.")
        print("1 - Войти в храм. \n2 - Зайти в хижину. \n3 - Спуститься обратно.")
        choice2 = input("Что ты сделаешь? (1/2/3): ")

        if choice2 == "1":
            print("\nВ храме ты находишь карту сокровищ. Ты следуешь за ней и находишь клад!")
            print("Поздравляю, ты стал богатым и знаменитым!")
        elif choice2 == "2":
            print("\nТы зашел в хижину и встретил старого мудреца. Он дал тебе совет, который помог тебе найти сокровище.")
            print("Поздравляю, ты победил!")
        else:
            print("\nТы решил не входить в храм и спустился обратно. Но сокровище так и осталось тайной.")
            print("Конец игры!")
    else:
        print("\nТы стоял слишком долго и волшебный лес исчез. Ты упустил свое приключение!")
        print("Конец игры!")

adventure_bot()
```
